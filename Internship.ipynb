```python
state_counts = intapp.groupby('Organization State').size().dropna()
state_counts = state_counts.head(20)
#again copied from above but I only want the top 20
```

```python

#asked chatgpt: "Is there a formula to make a heat map in python with plotly" and I filled in the blanks since this is way above my skill level

fig3 = go.Figure(data=go.Choropleth(
    locations=state_counts.index,   # state abbreviations like OH, CA, NY
    z=state_counts.values,          # counts
    locationmode='USA-states',      # tells Plotly these are US states
    colorscale='Blues',
    colorbar_title="Submissions"
))

fig3.update_layout(
    title_text='Internship Submissions by State',
    geo_scope='usa'  # limits map to U.S.

)
fig3.update_layout(
    title='Internships by State',
    geo=dict(
        lakecolor='rgb(255, 255, 255)'
    ))
fig3.show()
```

```python
#I learned that it's blank because plotly only recognizes state abreviations for maps, and 
# my spreadsheet has full state names so I pulled this from chatgpt to clean it up
us_state_abbrev = {
    'Alabama': 'AL','Alaska': 'AK','Arizona': 'AZ','Arkansas': 'AR',
    'California': 'CA','Colorado': 'CO','Connecticut': 'CT','Delaware': 'DE',
    'Florida': 'FL','Georgia': 'GA','Hawaii': 'HI','Idaho': 'ID',
    'Illinois': 'IL','Indiana': 'IN','Iowa': 'IA','Kansas': 'KS',
    'Kentucky': 'KY','Louisiana': 'LA','Maine': 'ME','Maryland': 'MD',
    'Massachusetts': 'MA','Michigan': 'MI','Minnesota': 'MN',
    'Mississippi': 'MS','Missouri': 'MO','Montana': 'MT',
    'Nebraska': 'NE','Nevada': 'NV','New Hampshire': 'NH',
    'New Jersey': 'NJ','New Mexico': 'NM','New York': 'NY',
    'North Carolina': 'NC','North Dakota': 'ND','Ohio': 'OH',
    'Oklahoma': 'OK','Oregon': 'OR','Pennsylvania': 'PA',
    'Rhode Island': 'RI','South Carolina': 'SC','South Dakota': 'SD',
    'Tennessee': 'TN','Texas': 'TX','Utah': 'UT','Vermont': 'VT',
    'Virginia': 'VA','Washington': 'WA','West Virginia': 'WV',
    'Wisconsin': 'WI','Wyoming': 'WY'
}

# clean + convert
intapp['Organization State'] = intapp['Organization State'].str.strip()
intapp['State Abbrev'] = intapp['Organization State'].map(us_state_abbrev)
```

```python
state_counts = intapp.groupby('State Abbrev').size().dropna()
```

```python
fig3 = go.Figure(data=go.Choropleth(
    locations=state_counts.index,   # state abbreviations like OH, CA, NY
    z=state_counts.values,          # counts
    locationmode='USA-states',      # tells Plotly these are US states
    colorscale='Blues',
    colorbar_title="Submissions"
))

fig3.update_layout(
    title_text='Internship Submissions by State',
    geo_scope='usa'  # limits map to U.S.

)
fig3.update_layout(
    title='Internships by State',
    geo=dict(
        lakecolor='rgb(255, 255, 255)'
    ))
fig3.show()
```
