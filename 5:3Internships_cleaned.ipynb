*I learned that it's blank because plotly only recognizes state abreviations for maps, and my spreadsheet has full state names so I pulled the code below from chatgpt to clean it up.*

```python

us_state_abbrev = {
    'Alabama': 'AL','Alaska': 'AK','Arizona': 'AZ','Arkansas': 'AR',
    'California': 'CA','Colorado': 'CO','Connecticut': 'CT','Delaware': 'DE',
    'Florida': 'FL','Georgia': 'GA','Hawaii': 'HI','Idaho': 'ID',
    'Illinois': 'IL','Indiana': 'IN','Iowa': 'IA','Kansas': 'KS',
    'Kentucky': 'KY','Louisiana': 'LA','Maine': 'ME','Maryland': 'MD',
    'Massachusetts': 'MA','Michigan': 'MI','Minnesota': 'MN',
    'Mississippi': 'MS','Missouri': 'MO','Montana': 'MT',
    'Nebraska': 'NE','Nevada': 'NV','New Hampshire': 'NH',
    'New Jersey': 'NJ','New Mexico': 'NM','New York': 'NY',
    'North Carolina': 'NC','North Dakota': 'ND','Ohio': 'OH',
    'Oklahoma': 'OK','Oregon': 'OR','Pennsylvania': 'PA',
    'Rhode Island': 'RI','South Carolina': 'SC','South Dakota': 'SD',
    'Tennessee': 'TN','Texas': 'TX','Utah': 'UT','Vermont': 'VT',
    'Virginia': 'VA','Washington': 'WA','West Virginia': 'WV',
    'Wisconsin': 'WI','Wyoming': 'WY'
}

# clean + convert
intapp['Organization State'] = intapp['Organization State'].str.strip()
intapp['State Abbrev'] = intapp['Organization State'].map(us_state_abbrev)
```

```python
state_counts = intapp.groupby('State Abbrev').size().dropna()
```

```python
#tyler helped me get figure out how to play around with this color so it showed the differences bettter 
bluescale = [
    [0.0, "#FFFFFF"],
    [0.05, "#E1CC94"],
    [0.1, "#E0B747"],
    [1, "#EFAB00"]
]
```

```python
fig7 = go.Figure(data=go.Choropleth(
    locations=state_counts.index,   
    z=state_counts.values,          
    locationmode='USA-states',      
    colorscale= bluescale, #"Blues",
    colorbar_title="Submissions"
))

fig7.update_layout(
    title={
        'text': 'Number of Internships by State Since Fall 2023',
        'x': 0.5,
        'xanchor': 'center'
    },
    geo=dict(
        scope='usa',
        lakecolor='rgb(255, 255, 255)'
    ),
        margin=dict(
        l=10,
        r=10,
        t=40,
        b=10
)
)
fig7.show()
#use this 
```

```python
#asked chatgpt how to zoom into ohio further, I had to install pgeocode at my terminal and import it 

nomi = pgeocode.Nominatim('us')

# I guess this makes the zip codes a string
intapp['Organization Zip'] = intapp['Organization Zip'].astype(str).str[:5]

# this gets the lat/lon so glad my professor had mercy on my soul and let me switch to this project
intapp['lat'] = intapp['Organization Zip'].apply(lambda x: nomi.query_postal_code(x).latitude)
intapp['lon'] = intapp['Organization Zip'].apply(lambda x: nomi.query_postal_code(x).longitude)
```

```python
ohio_data = intapp[intapp['Organization State'] == 'Ohio']
#getting ohio only
```

```python
zip_counts = ohio_data.groupby(['Organization Zip', 'lat', 'lon']).size().reset_index(name='count')
#grouping by zip codes
```

```python
#David helped me with this map from https://plotly.com/python/tile-map-layers/

fig8 = px.scatter_map(
    zip_counts,
    lat='lat',
    lon='lon',
    size='count',
    hover_name='Organization Zip'
    #scope='usa'
)
fig8.update_traces(marker=dict(color="#003976"))

fig8.update_layout(
    title='Internship Locations in Ohio (by ZIP Code)',
    geo=dict(
        center={'lat': 40.4173, 'lon': -82.9071},  # center of Ohio
        projection_scale=6  # zoom into Ohio
    )
)
fig8.update_layout(map_style="open-street-map")
fig8.update_layout(margin={"r":0,"t":0,"l":0,"b":0})

fig8.show()
```
