```python
#asked chatgpt how to zoom into ohio further, I had to install pgeocode at my terminal and import it 

nomi = pgeocode.Nominatim('us')

# I guess this makes the zip codes a string
intapp['Organization Zip'] = intapp['Organization Zip'].astype(str).str[:5]

# this gets the lat/lon so glad my professor had mercy on my soul and let me switch to this project
intapp['lat'] = intapp['Organization Zip'].apply(lambda x: nomi.query_postal_code(x).latitude)
intapp['lon'] = intapp['Organization Zip'].apply(lambda x: nomi.query_postal_code(x).longitude)
```

```python
ohio_data = intapp[intapp['Organization State'] == 'Ohio']
#getting ohio only
```

```python
zip_counts = ohio_data.groupby(['Organization Zip', 'lat', 'lon']).size().reset_index(name='count')
#grouping by zip codes
```

```python
#I don't love this, not useful. I am just not sure it's useful to anyone

fig4 = px.scatter_geo(
    zip_counts,
    lat='lat',
    lon='lon',
    size='count',
    hover_name='Organization Zip',
    scope='usa'
)

fig4.update_layout(
    title='Internship Locations in Ohio (by ZIP Code)',
    geo=dict(
        center={'lat': 40.4173, 'lon': -82.9071},  # center of Ohio
        projection_scale=6  # zoom into Ohio
    )
)

fig4.show()
```
