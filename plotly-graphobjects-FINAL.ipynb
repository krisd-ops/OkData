Right, so we're going to need to break these lists into their own columns, luckily using pandas there's already a function we know and love to do this quite easily: dataframe!

# Elle on the board draws a quick chart

There are 4 pokemon evolution stages: Starter Middle Final and Lone

Starter has next evolutions but no prev

Middle has both next and prev

Final has prev but no next

Lone has no prev no next

### So using numpy select we can make:

```python
conditions = [
    (pokedata['next_evolution'].apply(isinstance, args=(list,))) & (pokedata['prev_evolution'].apply(isinstance, args=(float,))),
    (pokedata['next_evolution'].apply(isinstance, args=(float,))) & (pokedata['prev_evolution'].apply(isinstance, args=(list,))),
    (pokedata['next_evolution'].apply(isinstance, args=(float,))) & (pokedata['prev_evolution'].apply(isinstance, args=(float,)))
]

choices = ["Starter", "Final", "Lone"]
pokedata['evolution_stage'] = np.select(conditions, choices, default="Middle")
#Donald
#Apply conditions; default to "Middle" if none are met
```

```python
#let's check our work!
pokedata.tail()
#perfect let's make a quick group so we can add an evolution visualization to our plot
```

```python
evol = pokedata.groupby('evolution_stage').count()['id']
evol
#Donald
```

```python
fig3 = make_subplots(rows = 1, cols = 2, specs=[[{'type':'domain'}, {'type':'domain'}]])
#using make_subplots here
fig3.add_trace(
    go.Pie(
        labels=typegroup.index, 
        values=typegroup.values,
        marker=dict(colors=pokecolors),
        title="Gen 1 Pokemon by Main Type"
        ),
        1, 1)
# you can copy and paste this from the other chart
fig3.add_trace(
    go.Pie(
        labels=evol.index, 
        values=evol.values,
        title="Gen 1 Pokemon by Evolution Stage"
        ),
        1, 2)

fig3.show()
#Donald stop
```

```python
# Cut if not enough time

traces = go.Scatter(x=pokedata['height_m'],
                    y=pokedata['weight_kg'],
                    mode='markers',
                    name='Lone')

layout = go.Layout(title=dict(text='Pokemon weight and height'),
    xaxis=dict(title='Weight (in kg)'),
    yaxis=dict(title='Height (in m)'),
    )

fig4 = go.Figure(data=[traces], layout=layout)

fig4.show()

#Now this plot is... deeply unhelpful, I basically don't know anything more than where I started,
# So I'm thinking, what if we broke up the pokemon into their evolutionary stages using multiple traces instead of 1?

#duplicate this Option+Shift+Down Arrow, or Alt+Shift+Down Arrow
```

```python
trace1 = go.Scatter(x=pokedata[pokedata['evolution_stage'] == 'Lone']['height_m'],
                    y=pokedata[pokedata['evolution_stage'] == 'Lone']['weight_kg'],
                    customdata=pokedata[pokedata['evolution_stage'] == 'Lone']['name'],

                    name='Lone')
# we're going to set up 4 traces or plots in this figure now, mostly so that we can do more fun things later
# copy and paste this i beg you, and if you can use Find and Replace Even Better

trace2 = go.Scatter(x=pokedata[pokedata['evolution_stage'] == 'Starter']['height_m'],
                    y=pokedata[pokedata['evolution_stage'] == 'Starter']['weight_kg'],
                    customdata=pokedata[pokedata['evolution_stage'] == 'Starter']['name'],

                    name='Starter')

trace3 = go.Scatter(x=pokedata[pokedata['evolution_stage'] == 'Middle']['height_m'],
                    y=pokedata[pokedata['evolution_stage'] == 'Middle']['weight_kg'],
                    customdata=pokedata[pokedata['evolution_stage'] == 'Middle']['name'],

                    name='Middle')
                    

trace4 = go.Scatter(x=pokedata[pokedata['evolution_stage'] == 'Final']['height_m'],
                    y=pokedata[pokedata['evolution_stage'] == 'Final']['weight_kg'],
                    customdata=pokedata[pokedata['evolution_stage'] == 'Final']['name'],

                    name='Final')


layout = go.Layout(title=dict(text='Pokemon weight and height by Evolution Stage'),
    xaxis=dict(title='Weight (in kg)'),
    yaxis=dict(title='Height (in m)'),
    )

fig5 = go.Figure(data=[trace1, trace2, trace3, trace4], layout=layout)

#dont do this bit immediately, then add it
fig5.update_traces(mode='markers',
    marker=dict(size=10, opacity=.5)
)

fig5.show()
```

# Wow don't we feel fancy?? 
But what if we were FANCIER, Let's add more information to our hover text, like the pokemon's name and their types!

Before, we broke up our type Lists into two columns... Now we are going to smush them Back together but into a **string!**

(For reasons that we'll get into when it's relevant)

Using numpy where - which is just says if a certain condition is met do something and if it isn't met do a different thing, super similar to numpy select from the evolution stages code earlier


```python
condition = pokedata['type1'].notna() & pokedata['type2'].notna()

pokedata['Types'] = np.where(condition, pokedata['type1'] + ' & ' + pokedata['type2'], pokedata['type1'])
pokedata.head()
```

```python
#Copy from earlier
trace1 = go.Scatter(x=pokedata[pokedata['evolution_stage'] == 'Lone']['height_m'],
                    y=pokedata[pokedata['evolution_stage'] == 'Lone']['weight_kg'],
                    #New bit! we're going to add custom data to say the name and text to say the type
                    #These are different because it is simpler to code and even though in the syntax it says text can be a list of strings it broke every single time and i couldn't find an example that used it
                    #custom data is quite literally just another variable i could use, there's like 3
                    customdata=pokedata[pokedata['evolution_stage'] == 'Lone']['name'],
                    text=pokedata[pokedata['evolution_stage'] == 'Lone']['Types'],

                    name='Lone')

trace2 = go.Scatter(x=pokedata[pokedata['evolution_stage'] == 'Starter']['height_m'],
                    y=pokedata[pokedata['evolution_stage'] == 'Starter']['weight_kg'],
                    customdata=pokedata[pokedata['evolution_stage'] == 'Starter']['name'],
                    text=pokedata[pokedata['evolution_stage'] == 'Starter']['Types'],

                    name='Starter')

trace3 = go.Scatter(x=pokedata[pokedata['evolution_stage'] == 'Middle']['height_m'],
                    y=pokedata[pokedata['evolution_stage'] == 'Middle']['weight_kg'],
                    customdata=pokedata[pokedata['evolution_stage'] == 'Middle']['name'],
                    text=pokedata[pokedata['evolution_stage'] == 'Middle']['Types'],

                    name='Middle')
                    

trace4 = go.Scatter(x=pokedata[pokedata['evolution_stage'] == 'Final']['height_m'],
                    y=pokedata[pokedata['evolution_stage'] == 'Final']['weight_kg'],
                    customdata=pokedata[pokedata['evolution_stage'] == 'Final']['name'],
                    text=pokedata[pokedata['evolution_stage'] == 'Final']['Types'],

                    name='Final')


layout = go.Layout(title=dict(text='Pokemon weight and height by Evolution Stage'),
    xaxis=dict(title='Weight (in kg)'),
    yaxis=dict(title='Height (in m)'),
    )

fig6 = go.Figure(data=[trace1, trace2, trace3, trace4], layout=layout)

#New bits! hovertemplate
fig6.update_traces(
    hovertemplate="<b>%{customdata}</b> "\
    "<br>%{text} Type"
    "<br>Weight: %{x}kg" \
    "<br>Height: %{y}m", 
    mode='markers',
    marker=dict(size=10, opacity=.5)
)

fig6.show()

#so now our scatterplot is a lot more useful because it follows the guideline of more information being discoverable, 
# what else might we add to add to the depth of the information present?
#Elle End
```
